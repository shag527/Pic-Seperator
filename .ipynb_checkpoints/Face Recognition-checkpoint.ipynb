## Face Recognition and Seperation

```python
# Importing libraries

import cv2
import face_recognition
import numpy as np
from os import listdir
from os.path import isfile,join
import shutil
import matplotlib.pyplot as plt
from time import time
import logging
import pylab as pl
from PIL import Image
import os
```

```python
# Loading Dataset Images

data_path='Training Samples/'

training_data,labels=[],[]
data=[]

i=0
for files in os.listdir(data_path): 
    i+=1
    image_path=data_path+files
    images=Image.open(image_path)
    images=images.resize((100,100))
    data.append(np.asarray(images,dtype=np.uint8).flatten())
    
    train_image=face_recognition.load_image_file(image_path)
    try:
        boxes = face_recognition.face_locations(train_image,model='cnn')
        train_encoding=face_recognition.face_encodings(train_image,boxes)
        training_data.append(train_encoding)
        labels.append('Gates')
        print('Image ',i,': loaded Successfully')
    except IndexError as e:
        print('Image ',i,': Face not Detected')
    labels.append((i))
```

```python
print("Total dataset size:")
print("n_samples: %d" % len(data))
```

```python
#Loading all the images to be matched

test_path='Source/'
Dest_path='Destination/'
test_files=[f for f in listdir(test_path) if isfile(join(test_path,f))]
```

```python
for image in test_files:
    test_image=face_recognition.load_image_file(test_path+image)
    try:
        test_location=face_recognition.face_locations(test_image,model='cnn')
        test_encoding=face_recognition.face_encodings(test_image,test_location)
        for encoded_face, located_face in zip(test_encoding,test_location):
            result=face_recognition.compare_faces(training_data,encoded_face,0.6)
            if True in result:
                print(image,'Matched')
                shutil.move(test_path+image,Dest_path+image)
            else:
                print(image,'Not Matched')
                
    except IndexError as e:
        print('Face not Detected')
```
